## **4. Numpy arange와 linspace 데이터 생성**

### **4-1. 연속적인 데이터 생성**

> - np.arange(): N 만큼 차이나는 숫자 생성
> - np.linspace(): N 등분한 숫자 생성

**<arange와 linspace 차이점>**


    * np.arange([시작값], 끝값, [스텝])
        ([]는 생략 가능)
        (끝값 포함하지 않음)

    * np.linspace(시작값, 끝값, 등분할숫자)
        (끝값 포함)
    



```python
# arrange()함수에 1개 값 입력
print(np.arange(3))

# arrange()함수에 2개 값 입력
print(np.arange(3, 7))

# arrange()함수에 3개 값 입력
# 10이상 30미만 까지 5씩 차이나게 생성
print(np.arange(10, 30, 5))
```

```python
# 0이상 2미만 까지 0.3씩 차이나게 생성
print(np.arange(0, 2, 0.3))
```

```python
# 0~99까지 100등분
x = np.linspace(0, 99, 100)
print(x)
print(x.size)  # 100
```

```python
# arange 함수는 끝 값을 포함하지 않기 때문에
# 0부터 1.25 미만까지(끝 값 포함 안하니까 1까지 출력) 0.25씩 차이나게 생성
print(np.arange(0, 1.25, 0.25))
```

```python
# 위와 동일한 결과를 linspace 함수를 이용하여 코딩
# 0부터 1까지(끝 값 포함) 5등분
print(np.linspace(0,1,5))
```

### **6-1. element wise 연산**
* 차원(축)을 기준으로 행렬 내에서 같은 위치에 있는 원소끼리 연산을 하는 방식

```python
# a와 b 배열 생성하여 출력
a = np.array([20,30,40,50])
b = np.arange(4)
print(a)
print(b)
```

### **6-2. 여러가지 곱셈**

- '*' : 각각의 원소끼리 곱셈 (Elementwise product, Hadamard product)
- '@' : 행렬 곱셈 (Matrix product)

```python
# A와 B 배열 생성하여 출력
A = np.array( [[1,1],
               [0,1]] )
B = np.array( [[2,0],
               [3,4]] )
print(A)
print(B)
```

```python
# A * B
# 각각의 원소끼리 곱셈
print(A * B)
```

```python
# A @ B
# 행렬 곱셈 사용
print(A @ B)
```

### **6-3. 자동 형 변환**
- 수치연산 진행할 때 각각의 .dtype이 다르면,
- 타입이 큰쪽(int < float < complex)으로 자동으로 변경 (UPCASTING)
 * complex = 허수를 포함함 복소수

```python
# a와 b 배열 생성 & 타입 확인
a = np.ones(3, dtype=np.int32)
b = np.linspace(0, np.pi,3)

print(a)
print(b)
print(a.dtype)
print(b.dtype)
```

```python
# a(int), b(float) 연산 시 float로 upcasting
c = a + b
print(c)
print(c.dtype)
```

### **6-4. 집계함수**

- .sum( ) : 합
- .min( ) : 최소값
- .max( ) : 최대값
- .argmax( ) : 최대값의 인덱스
- .cumsum( ) : 누적합 | 직전 요소를 누적으로 더한 배열을 반환

```python
# a 배열 생성 & 출력
# 0부터 8미만까지 출력하고 (2,4) 크기로 재가공하고 제곱하여 출력
a = np.arange(8).reshape(2, 4)**2
print(a)
```

```python
# 최대값
print(a.max())
```

```python
# 최대값의 인덱스
print(a)
print(a.argmax())
```

```python
# 모든 요소의 누적합
# 14 = 0 + 1 + 4 + 9
print(a)
print(a.cumsum())
```

```python
# numpy라이브러리를 사용하여 버블 차트 그리기

# 예제
a = np.arange(-5, 5)
print(a)
mask1 = abs(a) > 3
mask2 = abs(a) % 2 == 0

print(mask1)
print(mask2)
print()

print(a[mask1])
print(a[mask2])
print()

print(mask1 + mask2)
print(mask1 * mask2)
print(a[mask1 + mask2]) # 둘 중 하나의 조건이라도 참일 경우
print(a[mask1 * mask2]) # 두가지 조건이 모두 참일 경우

# 1000개의 랜덤 수를 추출해서 그 중 절대값이 50보다 큰 값을 걸러내어 버블 차트로 그려보자.
import matplotlib.pyplot as plt
import numpy as np
x = np.random.randint(-100, 100, 1000) # 1000개의 랜덤 수 추출
y = np.random.randint(-100, 100, 1000) # 1000개의 랜덤 수 추출
size = np.random.rand(len(x)) * 100    # 크기가 x데이터 개수만큼의 랜덤 수 추출하여 100 곱하기
# print(type(size))
# print(size)

mask1 = abs(x) > 50  # x 값 중 절대값이 50보다 큰 값을 걸러내기
mask2 = abs(y) > 50  # y 값 중 절대값이 50보다 큰 값을 걸러내기
x = x[mask1 + mask2]  # mask1 과 mask2중 하나라도 만족하는 값 저장
y = y[mask1 + mask2]
size = size[mask1 + mask2]
print(x.shape)
plt.scatter(x, y, s=size,  c=x, cmap='plasma', alpha=0.6)
plt.colorbar()
plt.show()
```

## **9. Numpy 크기 변경**
- .ravel : 1차원으로 변경
- .reshape : 지정한 차원으로 변경
- .T : 전치(Transpose) 변환: 행과 열을 바꾸기

```python
# a 배열 생성 & shape 출력
a = np.arange(12).reshape(3,4)
print(a)
print(a.shape)
```

```python
# .ravel : 모든 원소를 1차원으로 변경
print(a.ravel())
print(a.reshape(-1))
```

```python
# .T : [3,4]의 전치(transpose)변환으로 [4,3] 출력
print(a.T)
print(a.T.shape)
```

```python
# 만약 전치 형태로 a 배열에 저장하고 싶다면!
a = a.T
print(a)
```

## **10. Numpy 데이터 합치기**
- np.vstack() 와 np.hstack()를 통해 데이터를 합치기
- np.vstack(): axis=0(열) 기준으로 쌓음
- np.hstack(): axis=1(행) 기준으로 쌓음

image.png

## **11. Numpy 데이터 쪼개기**

- np.hsplit( )에 숫자N이 들어갈 경우, N등분 한다.
- np.hsplit( )에 리스트로 넣을 경우, axis=1 기준 인덱스로 데이터를 분할
- np.vsplit( )에 숫자N이 들어갈 경우, N등분 한다.

```python
# a 배열 생성 & 출력
a = np.arange(12).reshape(2, 6)
print(a)
```

```python
# [2,6] => [2,2] 데이터 3개로 등분
print(np.hsplit(a, 3))
```

```python
# [2,6] => [:, :3], [:, 3:4], [:, 4:]로 분할
# a를 3번째 열 ~ 4번째 열 미만 기준으로 분할하여 3개의 array를 반환
print(a)
print(np.hsplit(a, (3,4)))
```

```python
# arr배열을 2등분하여 출력
arr = np.array([1, 2, 3, 4, 5, 6]).reshape(2, 3)
print(arr)
print(np.vsplit(arr, 2))
```

```python
# NymPy를 이용하지 않은 경우
import matplotlib.pyplot as plt

t = []
p2 = []
p3 = []

for i in range (0, 50, 2) :
  t.append (i /10)
  p2.append((i / 10) ** 2)
  p3.append((i/10)**3)


plt.plot(t, t, 'r--', t, p2, 'bs', t, p3, 'g^')
plt.show()
```

```python
# NumPy를 이용해서 동일한 그래프 그리기

import matplotlib.pyplot as plt
import numpy as np

t = np.arange(0., 5., 0.2)
print(t)
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
a = np.random.rand(5) # 0 ~ 1 사이에 있는 5개의 랜덤함 실수 생성
print(a)
print(type(a))

print(np.random.choice(6, 10)) # 0 ~ 5 까지의 숫자를 랜덤하게 10번 선택

# 중복 없이
print(np.random.choice(10, 6, replace=False)) # 0 ~ 10 까지의 숫자를 랜덤하게 중복 없이 6번 선택

print(np.random.choice(6, 10, p=[0.1, 0.2, 0.3, 0.2, 0.1, 0.1])) # 0 ~ 5 까지의 숫자를 지정 확률로 랜덤하게 10번 선택

```

```python
import matplotlib.pyplot as plt
import numpy as np

dice = np.random.choice(6, 10)
print(dice)

plt.figure(figsize=(4, 3), dpi=100)
plt.hist(dice, bins=6)
plt.show()
```

```python
import matplotlib.pyplot as plt
import time
import random

startTime = time.time() # 초단위로 시간을 제공
print(startTime)

dice = []
for i in range(1000000) :
  dice.append(random.randint(1,6))

plt.hist(dice, bins=6)
plt.show()

endTime = time.time()
elapsedTime = endTime - startTime
print(elapsedTime, 'sec')
```

```python
import matplotlib.pyplot as plt
import numpy as np
import time

startTime = time.time() # 초단위로 시간을 제공
print(startTime)

dice = np.random.choice(6, 10000000)

plt.hist(dice, bins=6)
plt.show()

endTime = time.time()
elapsedTime = endTime - startTime
print(elapsedTime, 'sec')
```

```python
import csv
import matplotlib.pyplot as plt

file = open('/content/drive/MyDrive/Colab Notebooks/seoulPopulation.csv', 'r', encoding="utf-8")
data = csv.reader(file)

malePopList = []
femalePopList = []


for row in data :
  if '영등포' in row[0] :
    # print(row[3:104])
    for i in row[3:104] :
      malePopList.append(int(i.replace(",", "")))
    # print(row[106:207])
    for i in row[106:] :
      femalePopList.append(int(i.replace(",", "")))

# print(malePopList)
# print(len(malePopList))
# print(femalePopList)
# print(len(femalePopList))

plt.rc('font', family='NanumBarunGothic')
plt.rcParams['axes.unicode_minus'] = False # y축에 '-' 기호를 나타내기 위한 설정
plt.style.use('ggplot')
plt.title('영등포구 연령별 남녀 인구 분포')
plt.plot(malePopList, color='r',  alpha=0.5, label ='남성')
plt.plot(femalePopList, color='b',  alpha=0.5, label ='여성')
plt.legend()

plt.show()

file.close()


```

```python
import csv
import matplotlib.pyplot as plt

file = open('/content/drive/MyDrive/Colab Notebooks/seoulPopulation.csv', 'r', encoding="utf-8")
data = csv.reader(file)

malePopList = []
femalePopList = []
popGapList = []


for row in data :
  if '서초구' in row[0] :
    for i in row[3:104] :
      malePopList.append(int(i.replace(",", "")))
    for i in row[106:] :
      femalePopList.append(int(i.replace(",", "")))
    for i in range (3, 104) :
      popGapList.append(int(row[i].replace(",", "")) - int(row[i + 103].replace(",", "")))

# print(malePopList)
# print(len(malePopList))
# print(femalePopList)
# print(len(femalePopList))
# print(popGapList)
# print(len(popGapList))

plt.rc('font', family='NanumBarunGothic')
plt.rcParams['axes.unicode_minus'] = False # y축에 '-' 기호를 나타내기 위한 설정
plt.style.use('ggplot')
plt.title('서초구 연령별 남녀 인구 분포')
plt.plot(malePopList, color='r',  alpha=0.5, label ='남성')
plt.plot(femalePopList, color='b',  alpha=0.5, label ='여성')
plt.bar(range(101), popGapList)
plt.legend()

plt.show()

file.close()
```

```python
import csv
import matplotlib.pyplot as plt

file = open('/content/drive/MyDrive/Colab Notebooks/jejuPopulation.csv', 'r', encoding="utf-8")
data = csv.reader(file)

totalPopList = []
malePopList = []
femalePopList = []
popGapList = []


for row in data :
  if '5000000000' in row[0] :
    for i in row[3:104] :
      totalPopList.append(int(i.replace(",", "")))
    for i in row[106:207] :
      malePopList.append(int(i.replace(",", "")))
    for i in row[209:] :
      femalePopList.append(int(i.replace(",", "")))
    for i in range (106, 207) :
      popGapList.append(int(row[i].replace(",", "")) - int(row[i + 103].replace(",", "")) )

# print(totalPopList)
# print(len(totalPopList))
# print(malePopList)
# print(len(malePopList))
# print(femalePopList)
# print(len(femalePopList))
# print(popGapList)
# print(len(popGapList))
print(max(malePopList))
print(max(femalePopList))

plt.rc('font', family='NanumBarunGothic')
# plt.rcParams['axes.unicode_minus'] = False # y축에 '-' 기호를 나타내기 위한 설정
plt.style.use('ggplot')
plt.title('제주특별자치도 연령별 남녀 인구 분포')
plt.scatter(malePopList, femalePopList, c=range(101), cmap='jet')
plt.colorbar()
plt.xlabel('남성인구수')
plt.ylabel('여성인구수')
plt.plot(range(max(malePopList)), range(max(malePopList)))

plt.show()

file.close()
```
